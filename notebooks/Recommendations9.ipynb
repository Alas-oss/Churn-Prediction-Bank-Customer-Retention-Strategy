# Recommendations

## Which customers should the bank prioritize for retention?
### First and foremost, it should prioritize high-value, high-risk clients fitting the distinct behavioural and demographic profiles that we found through the models. Specifically:

+ The Single-Product Middle-Aged customers. The ones aged between 35-55, as the PDP showed that the likeliness quitouples from 10% to 50% in this window. And that subset of people becomes even more likely to churn when they hold only one bank product.

+ The Inactive High Net Worth Segment of customers. The bank should prioritize ones who's bank balances exceed $100,000 and have an active_member status of 0. The PDP showed that there is a permanent 3% risk escalation once a balance surpases the afformentioned sum, and SHAP showed that complete inactivity is the largest factor in customer churning.

+ Geographic Focus. It's additionally important to prioritize accounts based in Germany, as the Logistic Regression coefficients and SHAP global summary marked this as an independent and structural risk factor.

## What's the expected uplift from model-driven targeting vs random outreach?

As seen in the Customer Decile Segmentation Matrix:
- By targetting the top 25% highest-risk customers identified by the model, the bank focuses entirely on a group with a combined average churn rate of over 57%.

- Random outreach of 25% of the bank's customers results in a lower prevention rate of about to churn customers rather than when it's targeted.

- Model-driven customer targeting yields a x2.8 Lift Multiplier Score, compared to a random campaign with the same percentage of outreach. It maximizes the conversion of the retention budget and protects millions in asset balances.

# How can the bank implement an automated churn alert system?
The bank can use it to pipeline information by deploying the automated, batch-processing alert architecture by:
- Data Processing: At the end of every set time period (week(s) - month(s)), the bank can look through the active customer tables in the database and use the automated Python script on it. 
- Scoring Deciles: The matrix is passed through the automated model that has been trained and optimized, in this case CatBoost. Instead of just outputting a binary flag, the model can provide a percentagee (probability score) for every account.
- Filter & Action: To make sure that the bank can catch the customers that are about to churn, the system will use the specific threshold value of 0.30. That makes sure that we catch more customers that are at risk of churning. (As it's better to have a wider sweep that's within range of the resources, because getting a promotional email won't hurt a non-churning customer). So any account with a probability score higher than the customer threshold automatically triggers an alert flag.
- CRM Incorporation: The flagged accounts are instantly moved to the bank's CRM list. So the system will automatically segment into respective subgroups so that specific alerts go directly to them, for direct outreach, for the high-balance and high-risk customers. While standard alerts will be triggered as promotional emails, for customers who are either lower-risk or lower-balance.
