# Feature Engineering - Bank Customer Churn

```python
import numpy as np 
import pandas as pd 

df = pd.read_csv('../data/Bank Customer Churn Prediction.csv')

df.columns = df.columns.str.lower()
print(f"Available columns: {list(df.columns)}")
```

### Feature engineering Core

```python
# Segments tenure years into phases
df['tenure_group'] = pd.cut(
    df['tenure'], 
    bins=[-1, 2, 5, np.inf], 
    labels=['0-2 years', '3-5 years', '6+ years']
)

# In-case tenure = 0, so that we don't divide by 0
df['activity_rate'] = np.where(
    df['tenure'] == 0,
    df['active_member'],
    df['active_member']/df['tenure']
)

# The financial footprint of the customer
df['customer_value'] = df['balance'] * df['tenure']

# Service interaction history
df['service_interaction_freq'] = df['credit_card'] + df['active_member']

# A binary flag to show if someone is holding more than one ank product
df['is_multi_product_customer'] = np.where(df['products_number'] > 1, 1, 0)

# A proxy flag to check if someone's activity has dropped
df['recent_activity_drop'] = np.where(
    (df['active_member'] == 0) & (df['products_number'] == 1),
    1,
    0
)

print(f"New dataset shape: {df.shape}")
```
